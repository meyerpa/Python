```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

# Displaying array data as an image

The matplotlib command `imshow` displays array data as an image. In this notebook we'll take a first look at elevation data for the southern part of the Red River Valley.

```python
# Read the elevation data stored as a numpy data file.
# The value stored in the array is the elevation at a particular point, 
# in meters above sea level.

elevation = np.load("elevation_subset.npy")
```

An example of the `imshow` command is below. By default, `imshow` uses the minimum and maximum values in the array to decide how to scale the colors.

Run the cell to display the elevation data.

```python
# We are forcing this imshow command to use the old default colormap ('jet') for now.
plt.imshow(elevation, cmap='jet')
plt.colorbar()
```

## What went wrong? And how do you fix it?

Clearly the default choices `imshow` is making for this data are not useful. 

One way to try to fix it is to use the `vmin` and `vmax` arguments to `imshow`, which set which array values (elevation in this case) get set to the colors on the ends of the colorbar.

## Exercise 1

In the cell below try filling in a few values for vmin and vmax to see if you can bring out more detail.

```python
plt.figure(figsize=(12, 12))

plt.imshow(elevation, vmin=260, vmax=280, cmap='jet')
plt.colorbar()
```

# Image histograms

A better way to choose what minimum and maximum values to use is to plot a histogram of image values.

In this example, the histogram would display how many pixels have each elevation. The cell below illustrates the matplotlib command for making a histogram. Note that you can make a histogram that isn't on a log scale, but for images where most of the pixels will be about the same value a log display is more informative.

```python
plt.hist(elevation.flatten(), log=True, bins=100);
```

## Exercise 2

Use the histogram plot to choose an appropriate vmin and vmax for this data. You want to choose a vmin that is near the bottom of the range where "most" of the pixels are and a vmax near the top of that range.

In the cell below display the image using the vmin and vmax you choose.

```python
plt.figure(figsize=(12, 12))

plt.imshow(elevation, vmin=250, vmax=300, cmap='jet')
plt.colorbar()
```

## Excersie 2, continued

Although the range is from 0 to 600, the 0 and 600 values are essentially outliers. The rest of the data doesn't vary from 0 to 600 uniformly. Instead the majority of the data is from 250 to 300.

## Exercise 3: changing the colormap

You can change the colormap with the keyword argument `cmap` in `imshow`. in addition to ``viridis``, try these two colormaps: ``gray`` and ``inferno``
