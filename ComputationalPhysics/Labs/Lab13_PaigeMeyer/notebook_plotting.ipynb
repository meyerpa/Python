# Plotting in the notebook with `matplotlib`

There are two different was to make plots appear in a matplotlib notebook, one interactive, one not. 

Telling the notebook which of these you want to use is the only extra step you need to make a plot in a notebook compared to making a plot in a python script like you've been doing thus far.

To make static plots, put this near the top of the notebook: `%matplotlib inline` (the `%` tells the notebook this is a special notebook command, not plain python)

To make interactive plots, put this near the top of the notebook: `%matplotlib notebook`

You can switch between the two in a single notebook, but doing so may lead to odd behavior.

Note that you do **NOT** need to use `plt.show()` to display the plot.

We'll make a simple plot of a sine wave in both interfaces.

## Setup: imports and data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 6 * np.pi, num=200)
y = np.sin(x)
```

## Static plots: the inline interface

```python
plt.plot(x, y, label='sine function')
plt.title('Sine wave')
plt.legend()
```

### Changing the size of the figure

The cell below shows how to change the size of the figure. Note this does NOT affect the plot in any other cell, just in the cell below.

```python
plt.figure(figsize=(10, 1))

plt.plot(x, y, label='sine function')
plt.title('Sine wave')
plt.legend()
```

## Interactive plots: the notebook interface

```python
# you may have to execute this cell twice to get the notebook interface to work
# (see comments below for the reason why).

plt.figure()
plt.plot(x, y, label='sine function')
plt.title('Sine wave')
plt.legend()
```

# Instructions for the future

+ The `notebook` interface gets confused when there are a lot of plots open. For our purposes `inline` will often be better, but use `notebook` if you need to zoom in to a plot. 
+ The `notebook` interface works better if, in each cell of the notebook where you make a plot, you put `plt.figure()` before any other plotting commands. This ensures matplotlib makes a brand new figure for that cell.
+ Note that you can convert a `notebook` plot to a simple static plot by clicking the *off* button in the toolsbar above the plot.
+ Do **NOT** use `%matplotlib inline` and `%matplotlib notebook` in the same notebook like we did here. Pick one and put it at the top of the notebook with your imports. Put it *before* your import of matplotlib.
