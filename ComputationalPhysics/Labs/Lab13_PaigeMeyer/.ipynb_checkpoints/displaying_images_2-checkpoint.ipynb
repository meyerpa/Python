```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the elevation data stored as a numpy data file.
# The value stored in the array is the elevation at a particular point, 
# in meters above sea level.

elevation = np.load("elevation_subset.npy")
```

```python
plt.hist(elevation.flatten(), log=True, bins=100);
```

## Flag Colormap
This colormap highlights the differences in altitude. In particular, it shows differences in small altitudinal changes very well.

```python
plt.figure(figsize=(12, 12))

plt.imshow(elevation, vmin=260, vmax=280, cmap='flag')
plt.colorbar()
```

## Diverging colormap - BrBG
This color map does well in showing that there isn't any major jumps in elevation except where the river is. It tends to flow smoothly from one to another. Additionally, the Red River is distinct from the land.

```python
plt.figure(figsize=(12, 12))

plt.imshow(elevation, vmin=260, vmax=280, cmap='BrBG')
plt.colorbar()
```
