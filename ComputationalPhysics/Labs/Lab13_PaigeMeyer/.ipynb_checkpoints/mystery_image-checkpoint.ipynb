### Imports

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

### Get the image information

```python
# Read the elevation data stored as a numpy data file.
# The value stored in the array is the elevation at a particular point, 
# in meters above sea level.

img = np.load("mystery_image.npy")
```

### Find where data resides

```python
plt.hist(img.flatten(), log=True, bins=40);
np.min(img.flatten())
```

Notice that most of the data is between 2,950 and 4,000.

## Default Color Map

```python
plt.figure(figsize=(15, 15))

plt.imshow(img, vmin=3050, vmax=3100)
plt.colorbar()
```

## Inferno Color Map

```python
plt.figure(figsize=(15, 15))

plt.imshow(img, vmin=3050, vmax=3100, cmap= "inferno")
plt.colorbar()
```

## Blues Color Map

```python
plt.figure(figsize=(15, 15))

plt.imshow(img, vmin=3050, vmax=3100, cmap= "Blues")
plt.colorbar()
```

## Cool Color Map

```python
plt.figure(figsize=(15, 15))

plt.imshow(img, vmin=3050, vmax=3100, cmap= "cool")
plt.colorbar()
```

## Qualitative Color Map

```python
plt.figure(figsize=(15, 15))

plt.imshow(img, vmin=3050, vmax=3100, cmap= "Accent")
plt.colorbar()
```

## Flag Color Map

```python
plt.figure(figsize=(15, 15))

plt.imshow(img, vmin=3050, vmax=3100, cmap= "flag")
plt.colorbar()
```
