### Changing Limits

```python
from pylab import plot, ylim, xlabel, ylabel,show
from numpy import linspace,sin

x = linspace(0,10,100)
y = sin(x)
plot(x,y)
# change y limits
ylim(-1.1,1.1) 
# add axis labels
xlabel("x axis")
ylabel("y = sin x")
show()
```

### Changing lines

```python
from pylab import plot, ylim, xlabel, ylabel,show
from numpy import linspace,sin

x = linspace(0,10,100)
y = sin(x)
plot(x,y, "r--") # r for red
plot(x-1,y, "go") # do g for green
plot(x+1,y, "y-") # do y for yellow
plot(x+2,y, "ms") # do y for yellow
# colors include red(r), green(g), yellow(y), blue(b), cyan(c), magenta(m), black(k), white(w)
# line styles include solid line("-"), dashed line("--"), solid line("-"), square("s"), dots("o"), smaller dots(".")
ylim(-1.1,1.1) 
xlabel("x axis")
ylabel("y = sin x")
show()
```

### Plotting Experimental Data Excercise 3.1

```python
from numpy import loadtxt, linspace
from pylab import plot, show, ylim
import os

cwd = os.getcwd()
filename = os.path.join(cwd, "cpresources", "sunspots.txt")

data = loadtxt(filename, float)
x = data[:,0]
y = data[:,1]
#plot(x,y)

# take only first 1000 datapts
x = x[:1000]
y = y[:1000]
plot(x,y, "r")

# calculate the average of ten points
newy = []
for i in x:
    summ = 0 
    for j in linspace(i-5, i+5, 10):
        if j>=0 and j< len(x):
            summ += y[int(j)]
    newy.append(1/(2*5)*summ)
    
ylim(0, 200)
plot(x, newy, "c--")
show()
```

#### Wave Interference Example

```python
from math import sqrt, sin, pi, cos
from numpy import empty
from pylab import imshow, gray, show

wavelength = 5.0
k = 2*pi/wavelength
xi0 = 1.0
separation = 20.0      # sparation of centers in cm
side = 100.0           # side of the square in cm
points = 500           # number of grid points along each side
spacing  = side/points # spacing of points in cm

# Calculate the positions of the centers of the circles
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2

# make an array to store the heights
xi = empty([points, points], float)

# Calculate the values in the array
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*i
        r1 = sqrt((x-x1)**2+(y-y1)**2)
        r2 = sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*sin(k*r1) + xi0*sin(k*r2)
        
# Make the plot
imshow(xi, origin="lower", extent=[0,side,0,side])
gray()
show()
```
